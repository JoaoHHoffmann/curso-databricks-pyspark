### Criando DataFrame


Criando dataframe


```python
%python
df = spark.createDataFrame([("Pedro", 20), ("João", 21), ("José", 22)])
df.show()
# df.display()
```

Criar dataframe com schema

```python
schema = "ID Int, Nome String"
dados = [[1, "Pedro"], [2, "João"], [3, "José"]]
df1 = spark.createDataFrame(dados, schema)
df1.show()

# Criando view temporaria
df1.createOrReplaceTempView("df1")
```

```python
%sql
select * from df1
```

Transformação com spark

```python
from pyspark.sql.functions import sum
schema2 = "Produtos STRING, Vendas INT"
vendas = [["Caneta", 10], ["Lápis", 20], ["Caneta", 15]]

df3 = spark.createDataFrame(vendas, schema2)
df3.show()

agrupado = df3.groupBy("Produtos").agg(sum("Vendas").alias("Total"))
agrupado.show()
df3.createOrReplaceTempView("agrupado")

# Podemos concatenar as operações mesmo sem persistir
df3.groupBy("Produtos").agg(sum("Vendas").alias("Total")).show()
```

Selecionando colunas especificas

```python
df3.select("Produtos").show()
df3.select("Produtos", "Vendas").show()
```

Expressões e select

```python
from pyspark.sql.functions import expr
df3.select("Produtos", "Vendas", expr("Vendas * 0.2")).show()
```

Vendo o schema

```python
from pyspark.sql.types import StructType
schema = df3.schema
print(schema)
```
